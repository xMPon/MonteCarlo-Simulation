# Monte Carlo Pi Estimation\n
\n
This notebook runs a simple Monte Carlo simulation to estimate $\\pi$.

```python
# Set up import path and import the simulation function
from pathlib import Path
import sys

project_root = Path.cwd().parent  # Assume project root is parent directory
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))  # Add project root to sys.path if needed

from simulations import estimate_pi  # Import the pi estimation function
```

```python
# Run the Monte Carlo simulation to estimate pi
samples = 200000  # Number of random points
seed = 42  # Random seed for reproducibility
pi_hat = estimate_pi(samples=samples, seed=seed)  # Estimate pi
pi_hat  # Display the estimated value
```

```python
# Calculate the absolute error from the true value of pi
abs(pi_hat - 3.141592653589793)
```
